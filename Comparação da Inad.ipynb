```python
import pandas as pd
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
import numpy as np
import datetime
import plotly.express as px
```

```python
def consulta_bacen(codigo_bacen):
    url = 'http://api.bcb.gov.br/dados/serie/bcdata.sgs.{}/dados?formato=json'.format(codigo_bacen)
    df = pd.read_json(url)
    df['data'] = pd.to_datetime(df['data'], dayfirst=True)
    df.set_index('data', inplace=True)
    return df
```

```python
inad_mg = consulta_bacen(15937)
inad_es = consulta_bacen(15868)
inad_rj = consulta_bacen(15879)
inad_sp = consulta_bacen(15885)
```

```python
fig, ax=plt.subplots()
ax.plot(inad_mg, color='tab:blue', label='Inad_MG')
ax.plot(inad_es, color='tab:red', label='Inad_ES')
ax.plot(inad_rj, color='tab:cyan', label='Inad_RJ')
ax.plot(inad_sp, color='tab:green', label='Inad_SP')
plt.title('Comparativo Inadimplência de PF no Sudeste Brasileiro')
plt.legend()
ax.set_facecolor('white')
plt.show()
```
